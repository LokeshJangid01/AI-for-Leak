```python
import csv 
import pandas as pd
from io import StringIO
import matplotlib.pyplot as plt
import random


df = pd.read_csv('./final_merge.csv')
# print(df)

empty_rows = (df[df.isnull().any(axis=1)].index + 2).tolist()
# print(empty_rows)

graph_ranges = []
start = 2
        
# # Iterate through empty rows to create ranges
for empty_row in empty_rows:
    end = empty_row - 1
    graph_ranges.append((start, end))
    start = empty_row + 1

graph_ranges = graph_ranges[:random.randint(2,50)]

# Increase the size of the figure
plt.figure(figsize=(20, 8))

# # Define colors for each graph
colors = ['cyan', 'yellow', 'red', 'black', 'blue', 'green', 'pink', 'violet', 'magenta', 'grey']

# # Plot each graph with a different color
for i, (start, end) in enumerate(graph_ranges):
    graph_data = df.iloc[start:end]
    # # Determine the color for this graph
    # if 2 <= start <= 59599 and 2 <= end <= 59599:
    #     color_index = 0  # Same color for the desired range
    # else:
    color_index = (i + 1) % len(colors)  # Cycle through the colors for other ranges
    plt.plot(graph_data['Elapsed_Time'], graph_data['Pressure'],
             label=f'Graph {i+1}', color=colors[color_index], linewidth=0.5)
# Add labels and title
plt.xlabel('Time')
plt.ylabel('Pressure')
plt.title('Multiple Graphs')

# Add legend
plt.legend()

# Show theÂ plot
plt.show()

```
